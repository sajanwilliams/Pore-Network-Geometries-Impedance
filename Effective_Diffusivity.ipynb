```python
import numpy as np
import matplotlib.pyplot as plt

# Equation for effective diffusivity
k = np.linspace(0.1, 2, 100)
D = k/np.tanh(k)

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the line
ax.plot(k, D)

# Set x and y axes labels
ax.set_xlabel('Relative Pore Width')
ax.set_ylabel('Effective Diffusivity')

# Display the plot
plt.show()
```
