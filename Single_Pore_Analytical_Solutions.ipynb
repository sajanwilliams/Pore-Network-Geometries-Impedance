```python
# import library
import matplotlib.pyplot as plt
import numpy as np
```

**Infinite Pore**

```python
# Calculate real and imaginary parts of impedance for given frequencies
def imp(w):
    z = (1 - 1j) / ((2 * w) ** 0.5)
    x = z.real
    y = -z.imag
    return x, y

# Select frequencies
w = np.logspace(-2, 2, 100)
# Creating impedance values
x, y = imp(w)

# Plotting impedance
plt.scatter(x, y)
plt.ylabel('Imaginary')
plt.xlabel('Real')
plt.title('Infinite')
plt.show()
```

**Reflective Boundary Condition**

```python
#calculating real and imaginary parts of impedance for given frequencies
def imp(w,l):
    #impedance for reflective pore condition with only diffusion
    z = (np.tanh(l*(1j*w)**0.5))**(-1)/(1j*w)**0.5
    # extract real part
    x = np.real(z)
    # extract imaginary part
    y = -np.imag(z)
    return x, y

#select frequencies
w = np.logspace(-2, 2, 100)

#initialize "l" as 1
l = 15

#creating impedance values
x, y = imp(w,l)



# Plotting impedance
plt.scatter(x, y)
plt.ylabel('Imaginary')
plt.xlabel('Real')
plt.title('Reflective')
plt.show()
```

**Transmissive Boundary Condition**

```python
#calculating real and imaginary parts of impedance for given frequencies
def imp(w,l):
    #impedance for tranmis pore condition with only diffusion
    z = (np.tanh(l*(1j*w)**0.5))/(1j*w)**0.5
    # extract real part
    x = np.real(z)
    # extract imaginary part
    y = -np.imag(z)
    return x, y

#select frequencies
w = np.logspace(-4, 2, 100)

#initialize "l" as 1
l = 15

#creating impedance values
x, y = imp(w,l)


# Plotting impedance
plt.scatter(x, y)
plt.ylabel('Imaginary')
plt.xlabel('Real')
plt.title('Transmissive')
plt.show()
```
